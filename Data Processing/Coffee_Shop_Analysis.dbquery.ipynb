```sql
SELECT *
FROM workspace.default.coffee_shop_analysis
LIMIT 10;

SELECT 
      transaction_date AS purchase_date,
      Dayname(transaction_date) AS Day_name,
      Monthname(transaction_date) AS Month_name,
      DayofMonth(transaction_date) AS Day_of_month,

 CASE
      WHEN Dayname(transaction_date) IN ('Sun','Sat') THEN 'Weekend'
      ELSE 'Weekday'
 END AS Weekend_or_Weekday,

 CASE
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '00:00:00' AND '11:59:59' THEN 'Morning'
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '12:00:00' AND '17:59:59' THEN 'Afternoon'
      WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '18:00:00' AND '23:59:59' THEN 'Evening'
 END AS time_bucket,

      COUNT(DISTINCT transaction_id) AS Number_of_sales,
      COUNT(DISTINCT product_id) AS Number_of_products,
      COUNT(DISTINCT store_id) AS Number_of_stores,
      
      SUM(transaction_qty*unit_price) AS revenue_per_day,

-- Categorical columns
      store_location,
      product_category,
      product_detail
      
FROM workspace.default.coffee_shop_analysis
GROUP BY transaction_date,
         Dayname(transaction_date),
         Monthname(transaction_date),
         DayofMonth(transaction_date),
         CASE
              WHEN Dayname(transaction_date) IN ('Sun','Sat') THEN 'Weekend'
              ELSE 'Weekday'
         END,
         CASE
              WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '00:00:00' AND '11:59:59' THEN 'Morning'
              WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '12:00:00' AND '17:59:59' THEN 'Afternoon'
              WHEN date_format(transaction_time, 'HH:mm:ss') BETWEEN '18:00:00' AND '23:59:59' THEN 'Evening'
         END,
         store_location,
         product_category,
         product_detail
```
